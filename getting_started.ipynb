## FAQ

### Q: What is a persistent environment?
Whenever you turn on/off the Studio or add GPUs, the environment does not change.   
Anything you installed (pip, conda, etc...), your data and files will be there next time the Studio is live again!

### Q: How is this different than my laptop?
- This is on the cloud and your laptop is not. This means you can put data, code, checkpoints here and give collaborators access to these files.
- When you close your laptop, the code stops running. Here, the code keeps running even if you close the browser.   
- Coding is free on Studio because you get a free running CPU Studio. To code something else, stop the Studio and start a new one for free!   
- The Studio environment is permanent... so you know that every time you want to work on something, it will be there. On your laptop you have to setup environments all the time.
- A Studio can add up to 8 GPUs unlike your laptop. It also has unlimited storage.
